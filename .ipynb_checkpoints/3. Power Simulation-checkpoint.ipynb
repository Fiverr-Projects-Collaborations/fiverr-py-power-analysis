```python
import pandas as pd
from matplotlib import pyplot as plt

all_house_data = pd.DataFrame()

with open ('Data/allHouses.csv') as f:
    text_list = f.read().split("\n\n")
    for text in text_list:
        text = text.split("\n")


        # text = list(filter(lambda x: x != '', text))

        appliance_list = []
        power_rating = []
        usage = []

        for items in text:
            if 'Residents' in items:
                resident_info = items.split(":")[1]
            elif 'House No' in items:
                house_number = items.split(":")[1]
            else:
                appliance_list.append(items.split(':')[0])
                power_rating.append(items.split(':')[1])
                usage.append(items.split(':')[2])

        df = pd.DataFrame()
        df_usage = pd.DataFrame()

        df['appliance'] = appliance_list
        df['power_rating'] = power_rating
        df['usage'] = usage

        # print(df.head())

        df_usage = df['usage'].str.split(',', expand=True)
        usage_col = list(range(0, 24))
        df_usage = df_usage[usage_col].astype(str).astype(float)
        df.drop(columns=['usage'], inplace=True)
        dataset = pd.concat([df, df_usage], axis=1)

        dataset['sum'] = dataset[usage_col].sum(axis=1).astype(str).astype(float)
        dataset['no_of_resident'] = resident_info
        dataset['house_no'] = house_number

        # print(dataset)

        all_house_data = all_house_data.append(dataset)
        # print(all_house_data)

all_house_data
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2,figsize=(15,10))
fig.suptitle("Hourly Usage per Appliance",fontsize=16)
count=1
for i in range(0,2):
    for j in range(0,3):
        plt.gcf().autofmt_xdate()
        plt.tight_layout(pad=5)
        all_house_data[all_house_data.house_no==str(count)].groupby(['appliance'])['sum'].agg(sum).plot.bar(legend=True,ax=axes[j][i],label="Total Hourly usage")
        plt.gcf().autofmt_xdate()
        axes[j][i].set_title('House: '+ str(count))
        count=count+1
#axes[1].set_title('Power Consumption per appliance')
#all_house_data[all_house_data.house_no=="1"]
```

```python
# import seaborn as sns
# g = sns.FacetGrid(all_house_data, col="house_no", hue="appliance")
# g.map(plt.scatter, "power_rating","sum")
# g.add_legend()

all_house_data['consuption_per_year_kwh'] = (all_house_data['power_rating'].astype(float) * all_house_data['sum'].astype(float) * 365) / 1000
solar_price_kwh_usd = 0.085
elect_price_kwh_usd = 0.14
all_house_data['solar_price_per_year_usd'] = all_house_data['consuption_per_year_kwh'] * solar_price_kwh
all_house_data['elect_price_per_year_usd'] = all_house_data['consuption_per_year_kwh'] * elect_price_kwh
all_house_data
```

```python
ax = all_house_data.groupby('house_no').agg({'solar_price_per_year_usd':'sum','elect_price_per_year_usd':'sum'}).plot(legend=True)
ax.set_ylabel("Annual Electricity Price in USD")
ax.set_xlabel("House No")
```

```python
all_house_data['perc_savings'] = (all_house_data['elect_price_per_year_usd'] - all_house_data['solar_price_per_year_usd']) * 100/all_house_data['elect_price_per_year_usd']
all_house_data
```
