```python
import pandas as pd
from matplotlib import pyplot as plt

#Read CSV File as text
with open ('Data/myhouse.csv') as f:
    text_list = f.read().split("\n")

#Data cleaning and transformation
text_list = list(filter(lambda x: x != '', text_list))

appliance_list = []
power_rating = []
usage = []

for items in text_list:
    if 'Residents' in items:
        resident_info = items

    else:
        appliance_list.append(items.split(':')[0])
        power_rating.append(items.split(':')[1])
        usage.append(items.split(':')[2])


df = pd.DataFrame()
df_usage = pd.DataFrame()

df['appliance'] = appliance_list
df['power_rating'] = pd.to_numeric(power_rating)
df['usage'] = usage
df
```

```python
#Split hourly usage column to individual hours
df_usage = df['usage'].str.split(',',expand=True)
usage_col = list(range(0,24))
df_usage = df_usage[usage_col].astype(str).astype(float)
df.drop(columns=['usage'], inplace=True)
dataset = pd.concat([df, df_usage], axis=1)
#New column for total hourly usage
dataset['total_hourly_usage'] = dataset[usage_col].sum(axis=1).astype(str).astype(float)
dataset
```

```python
#Calculating power consumption per day for each appliance taking max power rating as base
dataset['power_consumption_per_day_kwh'] = dataset['power_rating'] * dataset['total_hourly_usage']/1000
dataset
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(15,4))
fig.suptitle("Power Usage per Appliance",fontsize=16)
dataset.groupby(['appliance'])['power_consumption_per_day_kwh'].agg(sum).plot(legend=True,ax=axes[1],label="Total Power Consumption")
dataset.groupby(['appliance'])['total_hourly_usage'].agg(sum).plot.barh(legend=True,ax=axes[0],label="Total Hourly Usage")
axes[0].set_title('Hourly Usage per appliance')
axes[0].set_xlabel("Cummulative hours")
axes[0].set_ylabel("Appliances")
axes[1].set_title('Power Consumption per appliance')
axes[1].set_xlabel("Appliances")
axes[1].set_ylabel("Poer Consumption in kWh")
```

```python
solar_price_kwh_usd = 0.085
elect_price_kwh_usd = 0.14
dataset['solar_price_per_year_usd'] = dataset['power_consumption_per_day_kwh'] * solar_price_kwh_usd * 365
dataset['elect_price_per_year_usd'] = dataset['power_consumption_per_day_kwh'] * elect_price_kwh_usd * 365
dataset
```

```python
ax = dataset.groupby('appliance').agg({'solar_price_per_year_usd':'sum','elect_price_per_year_usd':'sum'}).plot(legend=True)
ax.set_ylabel("Annual Electricity Price in USD")
ax.set_xlabel("Appliance")
ax.legend(['Solar Power','Electricity'])
```
